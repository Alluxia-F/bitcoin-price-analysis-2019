```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=Warning)

import plotly.express as px
```

### Part I 

Load Bitcoin Historical Price from S3

```python
bitcoin_2019=pd.read_csv('https://bitcoin-price-analysis.s3-us-west-2.amazonaws.com/BTC-USD.csv')
```

```python
bitcoin_2019=bitcoin_2019[::-1].reset_index(drop=True)
bitcoin_2019.head()
```

```python
bitcoin_2019['Daily_Average']=(bitcoin_2019['Open']+bitcoin_2019['High']+bitcoin_2019['Low']+bitcoin_2019['Close'])/4
```

```python
bitcoin_2019.head()
```

### Bitcoin 183 days & 365 days rolling back average

```python
bitcoin_2019['183days_rolling_back_average']=bitcoin_2019['Daily_Average'].rolling(183).mean().shift(-182)
bitcoin_2019['365days_rolling_back_average']=bitcoin_2019['Daily_Average'].rolling(365).mean().shift(-364)
bitcoin_2019
```

```python
bitcoin_2019.dropna(how='any',inplace=True)
bitcoin_2019
```

```python
bitcoin_rolling_price=bitcoin_2019[['Date','Daily_Average','183days_rolling_back_average','365days_rolling_back_average']]
bitcoin_rolling_price.head(30)
```

### Part II 
Load bitcoin historical price via coindesk API:

https://api.coindesk.com/v1/bpi/historical/close.json?start=2011-01-01&end=2019-11-25

Change the end date by yourself. 


```python
import json,urllib.request
data = urllib.request.urlopen("https://api.coindesk.com/v1/bpi/historical/close.json?start=2011-01-01&end=2019-11-26").read()
output = json.loads(data)
bitcoin_historical_price=pd.DataFrame([output['bpi']]).T 
bitcoin_historical_price.reset_index(inplace=True)
bitcoin_historical_price=bitcoin_historical_price.rename(columns={'index':'Date',0:'Daily_price'})
```

```python
bitcoin_historical=bitcoin_historical_price[::-1].reset_index(drop=True)
bitcoin_historical.head()
```

```python
bitcoin_historical['183days_rolling_back_average']=bitcoin_historical['Daily_price'].rolling(183).mean().shift(-182)
bitcoin_historical['365days_rolling_back_average']=bitcoin_historical['Daily_price'].rolling(365).mean().shift(-364)
```

```python
bitcoin_historical.head(20)
```

```python
bitcoin_historical_1 = bitcoin_historical.melt(id_vars=['Date']+list(bitcoin_historical.keys()[4:]), var_name='BITCOIN')
px.line(bitcoin_historical_1, x='Date', y='value', color='BITCOIN' )
```
