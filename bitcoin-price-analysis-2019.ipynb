### Part I 

Load Bitcoin Historical Price from S3

```python
bitcoin_2019=pd.read_csv('https://bitcoin-price-analysis.s3-us-west-2.amazonaws.com/BTC-USD.csv')
```

```python
bitcoin_2019=bitcoin_2019[::-1].reset_index(drop=True)
bitcoin_2019.head()
```

```python
bitcoin_2019['183days_rolling_back_average']=bitcoin_2019['Daily_Average'].rolling(183).mean().shift(-182)
bitcoin_2019['365days_rolling_back_average']=bitcoin_2019['Daily_Average'].rolling(365).mean().shift(-364)
bitcoin_2019
```

### Simple Trading Model : Buy Low & Sell High

The trading strategy is very simple: when daily price is lower than minimum of rolling 365 days average. BUY.

When the daily price is higher than manium of the rolling 365 days average. SELL.


```python
bitcoin_historical['365days-min']=bitcoin_historical['365days_rolling_back_average'].rolling(365).min().shift(-364)
bitcoin_historical['365days-max']=bitcoin_historical['365days_rolling_back_average'].rolling(365).max().shift(-364)
bitcoin_historical=bitcoin_historical.dropna(how='any')
```

```python
bitcoin_historical['buy']= np.where((bitcoin_historical['Daily_price']<bitcoin_historical['365days-min']),1,0)
bitcoin_historical['sell']= np.where((bitcoin_historical['Daily_price']>bitcoin_historical['365days-max']),1,0)
```

```python
# Jan. All buy
bitcoin_historical['cost']=bitcoin_historical['Daily_price']*bitcoin_historical['buy']
bitcoin_historical_2019=bitcoin_historical[:'2019-01-01']
bitcoin_historical_2019=bitcoin_historical_2019[::-1]
bitcoin_historical_2019['total_cost']=bitcoin_historical_2019['cost'].cumsum()
bitcoin_historical_2019.head(30)
```

```python
bitcoin_historical_2019[:'2019-05-27']
```

```python
### Sell All on May 26th when Daily Price > 365 days max. On May26th: the daily price is: $8738

total_sell=bitcoin_historical_2019[:'2019-05-27']['buy'].sum()*8738
total_cost=506404
profit=total_sell-total_cost
return_on_investment=profit/total_cost
str(return_on_investment*100)+'%'
```
